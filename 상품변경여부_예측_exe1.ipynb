```python
import pandas as pd

import matplotlib.pyplot as plt

change = pd.read_csv('A0004TC.csv')

sns.hist(change['상품변경여부'])

import seaborn as sns
sns.barplot(data=change, x='연령', y='총 사용기간(Month)')

change.isnull().sum()
change['~~'].fillna('0', inplace=True) #float타입은 0으로 변환
change['%%'].fillna(' ', inplace=True) #object타입은 ' '으로 변환

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
change['신규가입일자'] = le.fit_transform(change['신규가입일자'])
change['연계상품 여부'] = le.fit_transform(change['연계상품 여부'])
change['프로모션 정보'] = le.fit_transform(change['프로모션 정보'])
change['상품변경여부'] = le.fit_transform(change['상품변경여부'])
change.drop('일자', axis=1, inplace=True)

change.info()
#from sklearn.preprocessing import OneHotEncoder
change = pd.get_dummies(data=change, columns=['$$'], drop_first=True)

X = change.drop('상품변경여부', axis=1).values
y = change['상품변경여부'].values
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, max_features=9, max_depth=5, random_state=42)
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_valid)
from sklearn.metrics import Confusion_matrix
from sklearn.metrics import classification_report
cm = Confusion_matrix(y_pred, y_valid)
cm
sns.heatmap(cm, annot=True)
print(classification_report(y_valid, y_pred))

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
model = Sequential()
model.add(Dense(3, input_shape(10,), activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(4, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(4, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
es = EarlyStopping(monitor='val_loss', mode='min', patience=5, verbose=1)
cp = ModelCheckpoint(filepath='best_model.h5', monitor='val_loss', save_best_only=True, save_weights_only=True, verbose=1)
history = model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=20, batch_size=10, callbacks=[es, cp])

plt.plot(history.history['acc'])
plt.plot(history.history['loss'])
plt.plot(history.history['val_acc'])
plt.plot(history.history['val_loss'])
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Acc')
plt.legend(['acc', 'loss', 'val_acc', 'val_loss'])
plt.show()
```
